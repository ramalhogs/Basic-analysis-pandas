# Análise de clientes de um e-commerce

Este é um trabalho válido pela A2 da matéria de Introdução a Ciência de Dados, do curso de Ciência de Dados & Inteligência Artificial. De modo geral, o trabalho propõe uma visualização mais ampla de dados de uma empresa de eletrônicos em relação a venda de produtos aos seus clientes. A base de dados pode ser encontrada em: https://www.kaggle.com/prachi13/customer-analytics

```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df_comercio = pd.read_csv('ecommerce.csv')
```

```python
df_comercio.boxplot(column=['Weight_in_gms'], patch_artist=True,
                   boxprops=dict(facecolor='gray', color='black'),
                   capprops=dict(color='gray'),
                   whiskerprops=dict(color='gray'),
                   flierprops=dict(color='gray', markeredgecolor='gray'),
                   medianprops=dict(color='black'))
```

O boxplot nos da diversas informações sobre os pesos dos produtos. Com ele, agora sabemos o menor e maior valor, o primeiro e terceiro quartil e a mediana dos pesos.

```python
quantidade_descontos = df_comercio.Discount_offered.value_counts()

plt.bar(x=quantidade_descontos.index.values, height=quantidade_descontos.values,
        color=['blue'], edgecolor='black')
```

O gráfico informa a quantidade de descontos oferecidos pela empresa por dólar. Pelas informações geradas, podemos concluir que a maior concentração de descontos da empresa está entre 0 dólares e 10 dólares, havendo quantidades muito menores de descontos para valores acima disso.

```python
quantidade_produtos = df_comercio.Cost_of_the_Product.value_counts()

plt.bar(x=quantidade_produtos.index.values, height=quantidade_produtos.values,
        color=['green'])
```

Aqui, o gráfico mostra os custos de cada produto vendido. É possível notar que os produtos estão concentrados numa faixa de preços bem específica, não havendo outros muito baratos (menos que 100 dólares) ou muito maiores (mais que pouco mais de 300 dólares) O eixo x representa o preço do produto e o eixo y a contagem deles.

```python
quantidade_sexos = df_comercio.Gender.value_counts()
pct_homem = quantidade_sexos['M'] * 100 / quantidade_sexos.values.sum()
pct_mulher = quantidade_sexos['F'] * 100 / quantidade_sexos.values.sum()
print('Porcentagem de homens:', pct_homem) 
print('Porcentagem de mulheres:', pct_mulher)

plt.bar(x=quantidade_sexos.index.values, height=quantidade_sexos.values,
        color=['blue', 'red'], edgecolor='black')
```

Nesse gráfico é mostrado a quantidade de compras feitas por homem e por mulheres. Podemos chegar a conclusão de que as porcentagens de clientes do sexo masculino e do sexo feminino são bastante semelhantes, cada um com basicamente 50% do total.

```python
compras_anteriores = df_comercio.Prior_purchases.value_counts()

plt.bar(x=compras_anteriores.index.values, height=compras_anteriores.values,
        color=['brown'], edgecolor='black')
```

Aqui, foi exposto a quantidade de compras anteriores à atual de cada cliente. Podemos ver que a maioria possuia 3 compras, e uma quantidade pequena possui 7 ou mais.

```python
chegou_no_prazo = df_comercio['Reached.on.Time_Y.N'].value_counts()
print(chegou_no_prazo)

plt.bar(x=chegou_no_prazo.index.values, height=chegou_no_prazo.values,
        color=['blue'], edgecolor='black')
```

O gráfico acima mostra quantos produtos chegaram no prazo estimado e quantos não chegaram. A coluna marcada por 1 mostra os produtos que chegaram no tempo proposto, e a marcada por 0 não chegou.

```python
peso_e_desconto = df_comercio[['Discount_offered', 'Weight_in_gms']]
peso_e_desconto

#plt.bar(x=quantidade_sexos.index.values, height=quantidade_sexos.values,
#        color=['blue', 'red'], edgecolor='black')
#quantidade_sexos
plt.scatter(peso_e_desconto.Weight_in_gms.values, peso_e_desconto.Discount_offered.values,
            color='blue', edgecolor='black')
```

Aqui, os pontos relacionam o peso do produto com o desconto oferecido por ele. Estranhamente, dos mais de 10 mil casos analisados, apenas 6 produtos com mais de 4kg tiveram descontos acina de 10 dólares.

```python
df_comercio.boxplot(column=['Weight_in_gms'], by='Product_importance', patch_artist=True,
                   boxprops=dict(facecolor='gray', color='black'),
                   capprops=dict(color='gray'),
                   whiskerprops=dict(color='gray'),
                   flierprops=dict(color='gray', markeredgecolor='gray'),
                   medianprops=dict(color='black'))
```

Nesse gráfico, foi possível separar (por grau de importância dos produtos) 3 boxplot que nos dão informações sobre os pesos. Possivelmente o gráfico mostra que não há correlação entre peso e importância, pois as medidas ficaram bem parecidas, em sua grande maioria. Apenas o valor mínimo do peso de produtos de alta importância é maior, chegando a aproximadamente 3kg.
